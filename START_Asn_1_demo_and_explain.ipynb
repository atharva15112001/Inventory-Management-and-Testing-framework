#### Show Images

This is an example of how to display an image from a URL. 

```python
# This code will display an image given a URL
from PIL import Image
import requests
import pandas as pd


df = pd.read_csv('Strength_Training.csv')
url = df["image"][0]

im = Image.open(requests.get(url, stream=True).raw)
display(im)
```
